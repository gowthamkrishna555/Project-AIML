Types of Data 

Numerical data: This type of data consists of numerical values that can be measured, such as height, weight, temperature, and age.

Categorical data: This type of data consists of non-numerical values that are assigned to different categories, such as gender, race, or occupation.

Textual data: This type of data consists of unstructured text, such as articles, reviews, or social media posts.

Ordinal data: This type of data is similar to categorical data, but the categories have an inherent order or ranking, such as education level or job seniority.

Continuous data:This type of numerical data that can take on any value within a certain range. It is a type of quantitative data that can be measured on a continuous scale, such as length, time, weight, or temperature.

Heatmap:A heatmap is a graphical representation of data that uses color-coding to display the relative magnitude of values in a matrix. The values in the matrix are represented as a grid of squares, where each square is colored according to its value. The colors used in the heatmap can range from a single color gradient to a multi-color gradient, and they are typically chosen to highlight the differences in magnitude between values.

Pair plot:A pair plot, also known as a scatterplot matrix, is a type of data visualization that allows you to quickly and easily visualize the relationships between multiple variables in a dataset. It consists of a grid of scatterplots, where each scatterplot shows the relationship between two variables.In a pair plot, each variable is plotted against every other variable in the dataset, resulting in a grid of scatterplots. The diagonal of the grid shows a histogram or a density plot of each variable, which provides information about the distribution of the data.

Clustermap:A cluster map is a type of data visualization that uses clustering algorithms to group similar data points together and display them as clusters on a heatmap. Each row and column in the heatmap represents a data point, and the color of each cell in the heatmap indicates the similarity between the corresponding pair of data points.
